```python
# goal ATTGCTTAAC
# Steepest-Ascent hill climbing

import random
import string


def generate_random_seq(length=10):
    return [random.choice(string.printable) for _ in range(length)]
```

```python
def evaluate(seq):
    target = list("ATTGCTTAAC")
    diff = 0
    for i in range(len(target)):
        s = seq[i]
        t = target[i]
        diff+=abs(ord(s) - ord(t))  #h(n)
        
    return diff
        
```

```python
def mutate_sq(seq):
    index = random.randint(0, len(seq)-1)
    seq[index] = random.choice(string.printable)
    
    
    
```

```python
best = generate_random_seq()
best_score = evaluate(list(best))
```

```python
while True:
    print("best score so far", best_score,"Seq: ", "".join(best))
    
    if best_score == 0:
        break
        
    
    new_seq = list(best)
    mutate_sq(new_seq)
    new_score = evaluate(new_seq)
    
    if new_score < best_score:
        best_score = new_score
        best = new_seq
    
```
